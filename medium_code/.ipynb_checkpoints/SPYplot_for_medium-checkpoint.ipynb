```python
import numpy as np  
import pandas as pd   
import os
import matplotlib.pyplot as plt  
from matplotlib.dates import DateFormatter
```

```python
path = os.getcwd()
data_path  = os.path.dirname(path)+'/data/'
results_path = os.path.dirname(path)+'/results/'
```

```python
spy_file = 'SPY_2020+.csv' #Contains data from 12/30/2019 - 3/04/2021
vix_file = '^VIX_2020.csv' #Contains data from 1/2/2020 - 12/31/2020

spy_data = pd.read_csv(data_path+spy_file, index_col = 'Date')
spy_data.drop(['High','Open','Close','Low','Volume'],axis=1, inplace=True)

vix_data = pd.read_csv(data_path+vix_file, index_col = 'Date')
vix_data.drop(['Adj Close','High','Low','Close','Volume'],axis=1, inplace=True)

data = pd.merge(spy_data,vix_data,left_index=True,right_index=True,how='left')
data.rename(columns = {"Adj Close": "SPY"}, inplace=True)
data.rename(columns = {"Open": "VIX"}, inplace=True)
data = data.reset_index(level=['Date']);
data.Date = pd.to_datetime(data.Date, format='%Y-%m-%d')

len(data)
```

```python
t2 = data[(data.Date <= (data.Date[45] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[45])].Date
y2 = data[(data.Date <= (data.Date[45] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[45])].SPY
t3 = data[(data.Date <= (data.Date[100] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[100])].Date
y3 = data[(data.Date <= (data.Date[100] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[100])].SPY
t4 = data[(data.Date <= (data.Date[155] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[155])].Date
y4 = data[(data.Date <= (data.Date[155] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[155])].SPY
t5 = data[(data.Date <= (data.Date[210] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[210])].Date
y5 = data[(data.Date <= (data.Date[210] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[210])].SPY
t6 = data[(data.Date <= (data.Date[9] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[9])].Date
y6 = data[(data.Date <= (data.Date[9] + pd.Timedelta(days=45)))&
                               (data.Date>data.Date[9])].SPY
```

```python
fig = plt.figure(figsize=(16,6),dpi=100) 
ax1 = fig.add_subplot(1,1,1)

x = data[data['Date'].dt.year == 2020].Date
y1 = data[data['Date'].dt.year == 2020].SPY
#y2 = data[data['Date'].dt.year == 2020].VIX

date_form = DateFormatter("%m-%y")
ax1.xaxis.set_major_formatter(date_form)
ax1.plot(x, y1,color=(.6,.6,.6), linewidth=2, linestyle='-', label='SPY')
ax1.plot(t2,y2,color='k')
ax1.plot(t3,y3,color='g')
ax1.plot(t4,y4,color='b')
ax1.plot(t5,y5,color='orange')
ax1.plot(t6,y6,color='red')
ax1.legend(loc='upper left', frameon=True) 
plt.title('SPY: 2020',fontweight='bold');

plt.savefig(results_path+'sim_example.png',bbox_inches = 'tight',pad_inches=0.5,facecolor='w')
```
